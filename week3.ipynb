```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('housing_area_price.csv') # Corrected path to the uploaded CSV

# Display dataset
print(df)

# Plot scatter graph
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.show()

# Independent variable
X = df[['area']]

# Dependent variable
y = df['price']

# Create Linear Regression model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict price for 3300 sq ft
price_3300 = model.predict([[3300]])
print("Predicted price for 3300 sq ft:", price_3300)

# Model coefficient (slope)
print("Coefficient:", model.coef_)

# Model intercept
print("Intercept:", model.intercept_)

# Linear equation
print("Equation: Price =", model.coef_[0], "* Area +", model.intercept_)

# Manual calculation using equation
manual_price = 3300 * model.coef_[0] + model.intercept_
print("Manual prediction:", manual_price)

# Predict price for 5000 sq ft
price_5000 = model.predict([[5000]])
print("Predicted price for 5000 sq ft:", price_5000)
```
